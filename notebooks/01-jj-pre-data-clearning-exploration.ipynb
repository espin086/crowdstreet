```python
class PreCleanExplorationCSV():
    import pandas as pd
    
    def __init__(self, file):
        self.file = file
        return None
    
    def __str__(self):
        return "INFO: Contains a file called: {0}".format(self.file)
    
    def _load_file(self):
        df = pd.read_csv(self.file)
        return df
    
    def explore(self):
        df = self._load_file()
        results = {'datatypes': df.dtypes,
                   'head': df.head(),
                   'tail': df.tail(),
                   'shape': df.shape,
                   'info': df.info(),
                   'describe': df.describe(),

                  }
        print("""
        ---------------------------------------------
        Initial Exploration Report:
        ---------------------------------------------
        
        Datatype of DF:
        -----------------------
        {5}
        
        
        Head of DF:
        -----------------------
        {0}
        
        
        Tail of DF:
        -----------------------
        {1}
        
        
        Shape of DF:
        -----------------------
        {2}
        
        Info of DF:
        -----------------------
        {3}
        
        
        Describe DF:
        -----------------------
        {4}
        
        ---------------------------------------------
        END OF REPORT.
        ---------------------------------------------
        
        """.format(results['head'],results['tail'], results['shape'], results['info'], results['describe'], results['datatypes'] ))
        
        return results
        
    pass
        
    
```

```python
raw = PreCleanExplorationCSV("../data/raw/Offerings_2019-10-25.csv")
```

```python
print(raw)
```

```python
explore = raw.explore()
```
